## 1. Loading the data

Click on the codeblock `In []` below and press `Ctrl` + `Enter` to run the code. Below the codeblock you can see the output.

```python
from intro_functions import Intro
f = Intro() 
data = f.load_data('winequality-red.csv') 
```

Click on the next code block and press `Ctrl + Enter` again. Continue your way down towards the end of the script! 

```python
f.format_data(data)
data, new_data = f.divide_data(data)
f.add_demographics(data, new_data)
```

Look through the first 20 values of our dataset.

## 2. Set targets and features

We set a target: what we want to predict and features: the data we use for prediction.

```python
target = f.set_target(data, "quality")
features = f.set_features(data, "quality")
```

## 3. Training the model

Here, we fit our model parameters to our dataset. 

```python
my_model = f.train_model(features, target)
```

## 4. Predicting

Now, we make predictions for the quality of wine based on our `features`. 

```python
prediction = f.make_prediction(my_model, features)
```

## 5. Evaluating the model

We evaluate our model by looking how well it predicted the quality. We see how well our model does by counting the amount correct and incorrect predictions for quality. 

```python
evaluation = f.evaluate_prediction(target, prediction)
f.plot_evaluation_result(evaluation)
```

## 6. New data coming in! 

New data is coming in! The new dataset contains the most recent data of wine quality ratings. How is your model performing now? The piechart below shows how well our model is predicting based on the new data.

```python
new_target = f.set_target(new_data, "quality")
new_features = f.set_features(new_data, "quality")
```

```python
new_prediction = f.make_prediction(my_model, new_features)
new_evaluation = f.evaluate_prediction(new_target, new_prediction)
f.plot_evaluation_result(new_evaluation)
```
